# Assignment 2 — Spark DataFrames and Transformations

a) Theory
Transformations vs Actions in Spark

In Apache Spark, operations are divided into transformations and actions, forming the backbone of its execution model.

1. Transformations

Transformations are operations that create a new DataFrame/RDD from an existing one. They are lazy, meaning they are not executed immediately but only when an action is called.

Examples:

select() → Select specific columns
filter() → Apply conditions to data
groupBy() → Group data based on a column

Significance:

Enable lazy evaluation
Help Spark build an optimized DAG (Directed Acyclic Graph)
Reduce unnecessary computations
2. Actions

Actions trigger the execution of transformations and return results.

Examples:

show() → Displays data
collect() → Returns all data to driver
count() → Counts number of rows

Significance:

Execute the DAG created by transformations
Produce final output
Allow interaction with processed data

Key Difference

| Feature   | Transformations        | Actions                   |
| --------- | ---------------------- | ------------------------- |
| Execution | Lazy                   | Immediate                 |
| Purpose   | Build computation plan | Execute and return result |
| Output    | New DataFrame/RDD      | Value or display          |

## Step 1: Setup — Imports and SparkSession

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, count

spark = SparkSession.builder \
    .appName("Assignment2_DataFrame") \
    .master("local[*]") \
    .getOrCreate()

print(f"Spark Version: {spark.version}")
print("SparkSession ready!")
```

## Step 2: Create Sample Dataset

```python
data = [
    (1, "Alice", 25, 30000),
    (2, "Bob", 30, 40000),
    (3, "Charlie", 28, 35000),
    (4, "David", 35, 28000),
    (5, "Eva", 32, 50000)
]

columns = ["ID", "Name", "Age", "Salary"]
df = spark.createDataFrame(data, columns)

print("--- Full Dataset ---")
df.show()
print(f"Total rows: {df.count()} | Columns: {df.columns}")
```

## Step 3: Transformation — select()

```python
selected_df = df.select("Name", "Salary")
selected_df.show()
```

## Step 4: Transformation — filter()

```python
filtered_df = df.filter(col("Salary") > 30000)
filtered_df.show()
```

## Step 5: Transformation — groupBy() + count()

```python
grouped_df = df.groupBy("Age").count()
grouped_df.show()
```

## Step 6: Actions

```python
print("Display Data:")
df.show()

print("Total Records:", df.count())

names = df.select("Name").collect()
print("Employee Names:")
for row in names:
    print(row["Name"])
```

## Step 7: Stop Spark Session

```python
spark.stop()
print("SparkSession stopped.")
```
