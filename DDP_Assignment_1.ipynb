1. Distributed Data Processing

Distributed data processing divides large datasets into smaller parts and processes them across multiple machines (nodes) simultaneously.

Key features:

Parallel processing
Scalability
Fault tolerance
Data locality
2. Traditional Systems (Hadoop MapReduce)

Apache Hadoop processes data using disk storage.

Limitations:

High disk I/O
Slow performance
Not efficient for iterative tasks
3. Apache Spark Performance

Apache Spark improves performance through:

In-memory computation: Data stored in RAM → faster access
RDDs: Fault-tolerant distributed datasets
DAG execution: Optimized processing flow
Less disk usage: Reduced read/write operations
Faster processing: 10–100× quicker than Hadoop
4. Role of In-Memory Computation
Loads data once into memory
Reuses it multiple times
Reduces latency and speeds up processing
5. Conclusion

Spark is faster than traditional systems due to in-memory processing, optimized execution, and reduced disk operations, making it suitable for big data and real-time applications.

Write a PySpark program to:
- Create a DataFrame with at least 5 records (columns: ID, Name, Age, Salary)
- Display the schema
- Show all records
- Filter employees with salary greater than 30,000



```python
# Install PySpark - compatible with all Jupyter environments
import sys
import subprocess
subprocess.check_call([sys.executable, "-m", "pip", "install", "pyspark"])
print("PySpark installed successfully!")
```

## 1. Start Spark Session

```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DoubleType
from pyspark.sql.functions import col, desc

spark = SparkSession.builder \
    .appName("EmployeeApp") \
    .master("local[*]") \
    .getOrCreate()

print("Spark Version :", spark.version)
print("Session Status: Active")
```

## 2. Create DataFrame with 7 Employee Records

```python
schema = StructType([
    StructField("ID",     IntegerType(), False),
    StructField("Name",   StringType(),  False),
    StructField("Age",    IntegerType(), False),
    StructField("Salary", DoubleType(),  False)
])

data = [
    (1,  "Alice",   30, 45000.0),
    (2,  "Bob",     25, 28000.0),
    (3,  "Charlie", 35, 62000.0),
    (4,  "Diana",   28, 31000.0),
    (5,  "Edward",  40, 75000.0),
    (6,  "Fiona",   22, 22000.0),
    (7,  "George",  33, 53000.0)
]

df = spark.createDataFrame(data, schema=schema)
print("Total Records:", df.count())
```

## 3. Display Schema

```python
df.printSchema()
```

## 4. Show All Records

```python
df.show(truncate=False)
```

## 5. Filter - Salary Greater than 30,000

```python
filtered_df = df.filter(df["Salary"] > 30000)
filtered_df.show(truncate=False)
print(f"Employees with Salary > 30,000 : {filtered_df.count()}")
print(f"Employees with Salary <= 30,000: {df.count() - filtered_df.count()}")
```

## 6. Additional Operations

```python
# Statistical Summary
df.describe(["Age", "Salary"]).show()
```

```python
# Sort by Salary - Highest First
df.orderBy(desc("Salary")).show()
```

```python
# Filter by Age range 25 to 35
df.filter((df["Age"] >= 25) & (df["Age"] <= 35)).show()
```

```python
# Select specific columns
df.select("Name", "Salary").show()
```

## 7. Stop Spark Session

```python
spark.stop()
print("Spark Session Stopped.")
```
